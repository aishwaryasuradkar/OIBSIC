```python
columns=['Region', 'Date', 'Frequency', 'Estimated_Unemployment_Rate', 'Estimated_Employed', 'Estimated_Labour_Participation_Rate', 'Area']

df= pd.read_csv('Unemployment in India.csv', names=columns)
df.head(800)
```

VISUALIZATION OF DATA


```python
df.columns
```

```python
#used to plot graph between 2 numerical variables for example ageVSincome
sns.relplot(x='Estimated_Employed',y='Region',kind='line',data=df)
```

```python
df.columns
```

```python
#used to plot graph between one categorical variable(Gender on x-axis) and one or more numerical variable( Height corresponding to gender on y-axis)
sns.catplot(x='Estimated_Employed', y='Region', data=df)
```

```python
df.describe()
```

ANALYZING DATA BY BAR GRAPHS


```python
import plotly.express as px
fig1= px.bar(df, x='Region', y='Estimated_Unemployment_Rate', color='Region', title='Estimated_Unemployment_Rate', template='plotly')
fig1.update_layout(xaxis={'categoryorder': 'total descending'})
fig1.show()
```

UNEMPLOYMENT RATE (REGION WISE) BY BAR GRAPH

```python
fig2= px.bar(df, x='Region', y='Estimated_Unemployment_Rate', color= 'Region', title= 'Umemloyment rate (region wise) by BAR GRAPH', template='plotly')
fig2.update_layout(xaxis={'categoryorder': 'total descending'})
fig2.show()
```

UNPLOYMENT RATE BY BOX PLOT

```python
fig3 = px.box(df, x='Region', y='Estimated_Unemployment_Rate', title='Unemployment Rate by Region')
fig3.update_layout(xaxis={'categoryorder': 'total descending'})
fig3.show()
```

UNEMPLOYMENT RATE (STATEWISE) BY SCATTER PLOT

```python
fig4 =px.scatter(df, x='Region', y='Estimated_Unemployment_Rate', title='Unemployment Rate by Region')
fig4.update_layout(xaxis={'categoryorder': 'total descending'})
fig4.show()
```

ANALYZING DATA BY HISTOGRAM

```python
fig5 =px.histogram(df, x='Region', y='Estimated_Unemployment_Rate', title='Unemployment Rate by Region')
fig5.update_layout(xaxis={'categoryorder': 'total descending'})
fig5.show()
```
